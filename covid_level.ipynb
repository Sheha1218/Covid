```python
import opendatasets as od
import pandas as pd
import numpy as np
```

```python
od.download('https://www.kaggle.com/datasets/iamhungundji/covid19-symptoms-checker')
```

```python
data=pd.read_csv('/content/covid19-symptoms-checker/Cleaned-Data.csv')
```

```python
data=data.drop(['Country'],axis=1)
```

```python
abs(data.corr()[['Severity_Mild','Severity_Moderate','Severity_None','Severity_Severe']])
```

```python
x=data.drop(['Severity_Mild','Severity_Moderate','Severity_None','Severity_Severe'],axis=1)
```

```python
y=data[['Severity_Mild','Severity_Moderate','Severity_None','Severity_Severe']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import pickle
```

```python
with open("level_model.pkl", "wb") as file:
    pickle.dump(model, file)
```
